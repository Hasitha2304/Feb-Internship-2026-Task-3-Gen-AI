```python
# Install libraries

!pip install -U langchain transformers accelerate langchain-community


# Imports

import os
from langchain_core.prompts import PromptTemplate
from langchain_community.llms import HuggingFacePipeline
from transformers import pipeline


# LangSmith setup

os.environ["LANGCHAIN_TRACING_V2"] = "true"
os.environ["LANGCHAIN_API_KEY"] = input("Enter LangSmith API Key: ")


# Load HuggingFace model

pipe = pipeline(
    "text-generation",
    model="gpt2",
    max_new_tokens=100,
    truncation=True,
    pad_token_id=50256
)

llm = HuggingFacePipeline(pipeline=pipe)


# Dataset

job_description = """
Looking for a Data Scientist with:
Python, Machine Learning, Deep Learning,
Pandas, NumPy, Scikit-learn,
NLP, Data Visualization,
Minimum 2 years experience
"""

resumes = {
    "strong": "Python Machine Learning Deep Learning NLP Pandas NumPy Scikit-learn 3 years experience",
    "average": "Python Pandas Data Visualization 1 year experience",
    "weak": "MS Word Communication no experience"
}


# Task 1 Skill Extraction

extract_prompt = PromptTemplate.from_template(
"Extract skills, tools, and experience from: {resume}"
)

def extract_info(resume):
    return llm.invoke(extract_prompt.format(resume=resume))


# Task 2 Matching

match_prompt = PromptTemplate.from_template(
"Compare resume with job description and list matching and missing skills: {resume} {job}"
)

def match_resume(resume, job):
    return llm.invoke(match_prompt.format(resume=resume, job=job))


# Task 3 Scoring

score_prompt = PromptTemplate.from_template(
"Give a score from 0 to 100 based only on matching: {text}"
)

def score_candidate(text):
    return llm.invoke(score_prompt.format(text=text))


# Task 4 Explanation

explain_prompt = PromptTemplate.from_template(
"Explain why this candidate is suitable or not based on the job description: {resume} {job}"
)

def explain_score(resume, job):
    return llm.invoke(explain_prompt.format(resume=resume, job=job))


# Full Pipeline

def run_pipeline(name, resume, job):
    print("\n-----------------------------")
    print(f"{name.upper()} CANDIDATE")
    print("-----------------------------")

    extracted = extract_info(resume)
    print("\nExtracted:\n", extracted)

    match = match_resume(resume, job)
    print("\nMatch:\n", match)

    score = score_candidate(match)
    print("\nScore:\n", score)

    explanation = explain_score(resume, job)
    print("\nExplanation:\n", explanation)


# Run all candidates

for key in resumes:
    run_pipeline(key, resumes[key], job_description)
```
